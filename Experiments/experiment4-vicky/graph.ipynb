```python
import re
def extract(filename, approx_kl, entropy_loss, explained_variance, loss, policy_gradient_loss, value_loss, iterations):
    with open(filename, "r") as f:
        lines = f.readlines()

    for i in range(iterations):
        # Calculate the line index for the current iteration
        addition = (18 * i) + 7

        # Check if the calculated index exceeds the number of lines in the file
        if addition + 10 > len(lines):  # Using +10 as a buffer to ensure all data lines are present
            break

        # Updated regular expressions to correctly parse scientific notation
        sci_num_regex = r'-?\d+(?:\.\d+)?(?:e-?\d+)?'

        # Extract and append the values for each metric
        approx_kl.append(float(re.findall(sci_num_regex, lines[7 + addition])[0]))
        entropy_loss.append(float(re.findall(sci_num_regex, lines[10 + addition])[0]))
        explained_variance.append(float(re.findall(sci_num_regex, lines[11 + addition])[0]))
        loss.append(float(re.findall(sci_num_regex, lines[13 + addition])[0]))
        policy_gradient_loss.append(float(re.findall(sci_num_regex, lines[15 + addition])[0]))
        value_loss.append(float(re.findall(sci_num_regex, lines[16 + addition])[0]))
```

```python
#Supply filenames
filename,approx_kl,entropy_loss,explained_variance,loss,policy_gradient_loss,value_loss = './data2-4frame.txt',[],[],[],[],[],[]
filename2,approx_kl2,entropy_loss2,explained_variance2,loss2,policy_gradient_loss2,value_loss2 = './data3-3frame.txt',[],[],[],[],[],[]
filename3,approx_kl3,entropy_loss3,explained_variance3,loss3,policy_gradient_loss3,value_loss3 = './data4-2frame.txt',[],[],[],[],[],[]
#Extract data
extract(filename,approx_kl,entropy_loss,explained_variance,loss,policy_gradient_loss,value_loss,49)
extract(filename2,approx_kl2,entropy_loss2,explained_variance2,loss2,policy_gradient_loss2,value_loss2,49)
extract(filename3,approx_kl3,entropy_loss3,explained_variance3,loss3,policy_gradient_loss3,value_loss3,49)

# 检查每个列表的长度
(len(approx_kl3), len(entropy_loss3), len(explained_variance3), len(loss3), len(policy_gradient_loss3), len(value_loss3))
```
